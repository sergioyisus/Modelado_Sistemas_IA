# **Funciones de activación**

Es una función que introduce no linealidad en el modelo.
Sin funciones de activación, una red neuronal, sin importar cuántas capas tenga, se comportaría como una simple combinación lineal de sus entradas. Las funciones de activación permiten a las redes neuronales aprender y modelar relaciones más complejas y no lineales en los datos.

```python
# Importamos librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# # Datos genericos para las funciones
x = np.linspace(-10, 10, 400)
x_leaky = np.where(x < 0, 0.01 * x, x)
b = 0
alpha = 1.0
lambda_selu = 1.0507
```

**Lineal**

$y=x+b$

Se utiliza principalmente en la capa de salida para problemas de regresión. Al no aplicar ninguna transformación no lineal significativa, permite que la salida tome cualquier valor real, lo que es útil para predecir valores continuos.

Contextos: Problemas de regresión, algunas capas intermedias de redes neuronales donde no se requiere una transformación no lineal.







```python
# formula
y = x + b

# Grafica Lineal
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Lineal')
plt.title('Función de Activación Lineal')
plt.grid(True)
```

**Sigmoid (Logística)**

$\sigma(x) = \frac{1}{1 + e^{-x}}$

Se utiliza principalmente en la capa de salida para problemas de clasificación binaria. Su rango es de 0 a 1, lo que facilita la interpretación de los resultados como probabilidades.

Contextos: Clasificación binaria, redes neuronales simples.

```python
# Formula
sigmoid = 1 / (1 + np.exp(-x))

# Gráfica Sigmoid
plt.figure(figsize=(8, 6))
plt.plot(x, sigmoid, label='Sigmoid')
plt.title('Función de Activación Sigmoid')
plt.grid(True)
```

**Tanh (Tangente hiperbólica)**

$\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$

Similar a la función Sigmoid, pero mapea los valores de entrada entre -1 y 1, lo que puede ser beneficioso en capas ocultas al centrar los datos alrededor de cero.

Contextos: Redes neuronales recurrentes (RNN), capas ocultas de redes neuronales.


```python
# Formula
tanh = np.tanh(x)

# Gráfica Tanh
plt.figure(figsize=(8, 6))
plt.plot(x, tanh, label='Tanh', color='orange')
plt.title('Tanh')
plt.grid(True)
```

**ReLU (Unidad Lineal Rectificada)**

$\text{ReLU}(x) = \max(0, x)$

Es la función de activación más utilizada en las capas ocultas de redes neuronales profundas debido a su simplicidad y eficiencia en el entrenamiento.

Contextos: Redes neuronales convolucionales (CNN), capas ocultas de redes neuronales profundas en general.

```python
# Formula
relu = np.maximum(0, x)

# Gráfica Relu
plt.figure(figsize=(8, 6))
plt.plot(x, relu, label='ReLU', color='green')
plt.title('ReLU')
plt.grid(True)
```

**Leaky ReLU**

$\text{Leaky ReLU}(x) =
\begin{cases}
x & \text{si } x \geq 0 \\
\alpha x & \text{si } x < 0
\end{cases}$

Es una variante de ReLU que permite pequeños valores negativos, lo que puede ayudar a resolver el problema de las "unidades muertas" en ReLU.

Contextos: Redes neuronales profundas donde ReLU podría no funcionar bien.


```python
# Formula
leaky_relu = np.maximum(0.01 * x, x)

# Gráfica Leaky ReLU
plt.figure(figsize=(8, 6))
plt.plot(x, leaky_relu, label='Leaky ReLU', color='red')
plt.title('Leaky ReLU')
plt.grid(True)
```

**Softmax**

$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$





Se utiliza en la capa de salida para problemas de clasificación multiclase. Convierte los valores en una distribución de probabilidad.

Contextos: Clasificación multiclase, capas de salida de redes neuronales.

```python
# Formula
softmax = np.exp(x) / np.sum(np.exp(x))

# Gráfica Softmax
plt.figure(figsize=(8, 6))
plt.plot(x, softmax, label='Softmax', color='purple')
plt.title('Softmax')
plt.grid(True)
```

**ELU (Exponential Linear Unit)**

Ayuda a mitigar el problema de la "muerte de ReLU" y mejora la convergencia del modelo.

Contextos: Redes neuronales profundas donde se necesita una mejor convergencia y manejo de valores negativos.

```python
elu = np.where(x >= 0, x, alpha * (np.exp(x) - 1))

plt.figure(figsize=(8, 6))
plt.plot(x, elu, label='ELU', color='blue')
plt.title('ELU')
plt.grid(True)
```

**SELU (Scaled Exponential Linear Unit)**

Permite la autocorrección y autorregulación de la activación de las neuronas, mejorando la estabilidad del modelo.
Contextos: Redes neuronales profundas con muchas capas.

```python
selu = lambda_selu * np.where(x >= 0, x, alpha * (np.exp(x) - 1))

plt.figure(figsize=(8, 6))
plt.plot(x, selu, label='SELU', color='orange')
plt.title('SELU')
plt.grid(True)
```

**Swish**

Introduce una no linealidad suave que mejora el rendimiento en ciertos casos.

Contextos: Variantes de redes neuronales profundas, como en arquitecturas avanzadas de redes convolucionales.

```python
swish = x / (1 + np.exp(-x))

plt.figure(figsize=(8, 6))
plt.plot(x, swish, label='Swish', color='green')
plt.title('Swish')
plt.grid(True)
```

**Mish**

Ofrece una combinación de suavidad y robustez, mejorando el rendimiento en tareas específicas.

Contextos: Redes neuronales convolucionales y arquitecturas profundas.

```python
mish = x * np.tanh(np.log1p(np.exp(x)))

plt.figure(figsize=(8, 6))
plt.plot(x, mish, label='Mish', color='red')
plt.title('Mish')
plt.grid(True)
```

**Sigmoid:** Suaviza los valores de salida en el rango [0, 1], utilizada principalmente para problemas de clasificación binaria.

**Tanh:** Similar a la Sigmoid pero mapea los valores en el rango [-1, 1], utilizada en capas ocultas.

**ReLU:** Activa las neuronas solo si la entrada es positiva, ampliamente usada en capas ocultas de redes neuronales profundas.

**Leaky ReLU:** Variante de ReLU que permite valores negativos pequeños, ayuda a evitar unidades muertas.

**Softmax:** Convierte los valores en probabilidades, usada en la capa de salida para problemas de clasificación multiclase.

**ELU (Exponential Linear Unit):** Ayuda a mejorar la convergencia y manejar valores negativos, evitando el problema de la "muerte de ReLU".

**SELU (Scaled Exponential Linear Unit):** Permite la autocorrección y autorregulación de la activación de las neuronas, mejorando la estabilidad del modelo.

**Swish:** Introduce una no linealidad suave, mejorando el rendimiento en ciertos casos.

**Mish:** Ofrece una combinación de suavidad y robustez, mejorando el rendimiento en tareas específicas.

```python
# Configuramos el gráfico
plt.figure(figsize=(12, 10))

# Sigmoid
plt.subplot(3, 3, 1)
plt.plot(x, sigmoid, label='Sigmoid')
plt.title('Sigmoid')
plt.grid(True)

# Tanh
plt.subplot(3, 3, 2)
plt.plot(x, tanh, label='Tanh', color='orange')
plt.title('Tanh')
plt.grid(True)

# ReLU
plt.subplot(3, 3, 3)
plt.plot(x, relu, label='ReLU', color='green')
plt.title('ReLU')
plt.grid(True)

# Leaky ReLU
plt.subplot(3, 3, 4)
plt.plot(x, leaky_relu, label='Leaky ReLU', color='red')
plt.title('Leaky ReLU')
plt.grid(True)

# Softmax
plt.subplot(3, 3, 5)
plt.plot(x, softmax, label='Softmax', color='purple')
plt.title('Softmax')
plt.grid(True)

# ELU
plt.subplot(3, 3, 6)
plt.plot(x, elu, label='ELU', color='blue')
plt.title('ELU')
plt.grid(True)

# SELU
plt.subplot(3, 3, 7)
plt.plot(x, selu, label='SELU', color='orange')
plt.title('SELU')
plt.grid(True)

# Swish
plt.subplot(3, 3, 8)
plt.plot(x, swish, label='Swish', color='green')
plt.title('Swish')
plt.grid(True)

# Mish
plt.subplot(3, 3, 9)
plt.plot(x, mish, label='Mish', color='red')
plt.title('Mish')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Vamos a destacar algunas diferencias importantes:

**ELU vs. ReLU:**

ELU introduce una suavidad para los valores negativos, lo que puede mejorar la convergencia y el rendimiento del modelo, especialmente en capas profundas.
ReLU simplemente aplana los valores negativos a cero, lo que puede llevar al problema de unidades muertas donde las neuronas dejan de aprender si quedan atrapadas en la región negativa.

**SELU vs. ELU:**

SELU no solo proporciona suavidad para los valores negativos sino que también escala las activaciones de manera que la media y la varianza se mantienen constantes, lo que es beneficioso para la normalización interna del flujo de datos en redes profundas.
ELU no tiene esta propiedad de escalado automático.

**Swish vs. Sigmoid:**

Swish introduce una multiplicación de
𝑥 con la Sigmoid, permitiendo una activación no lineal más suave que puede ayudar a mejorar el flujo de gradientes durante el entrenamiento.
La función Sigmoid sola tiende a saturar para valores muy positivos o muy negativos, lo que puede dificultar el aprendizaje.

**Mish vs. ReLU/Swish:**

Mish combina características de Swish y Tanh, proporcionando una activación suave que puede mejorar tanto la robustez como el rendimiento del modelo en comparación con ReLU o incluso Swish en ciertos contextos.

```python
# Graficas
plt.figure(figsize=(12, 8))

# Sigmoid
plt.plot(x, sigmoid, label='Sigmoid', linestyle='--')
# ReLU
plt.plot(x, relu, label='ReLU', linestyle='--')
# ELU
plt.plot(x, elu, label='ELU')
# SELU
plt.plot(x, selu, label='SELU')
# Swish
plt.plot(x, swish, label='Swish')
# Mish
plt.plot(x, mish, label='Mish')

plt.title('Comparación entre Funciones de Activación')
plt.legend()
plt.grid(True)
plt.show()
```



**Sigmoid (línea punteada):** Suaviza los valores de salida entre 0 y 1.

**ReLU (línea punteada):** Activa las neuronas solo si la entrada es positiva.

**ELU:** Suaviza los valores negativos, evitando unidades muertas y mejorando la convergencia.

**SELU:** Similar a ELU, pero con escalado automático para mantener la media y varianza constantes.

**Swish:** Combina las propiedades de Sigmoid y linealidad, proporcionando una activación más suave.

**Mish**: Combina características de Swish y Tanh, ofreciendo una activación suave y robusta.
